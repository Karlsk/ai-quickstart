````python
# å¯¼å…¥å¿…è¦çš„æ¨¡å—
import os, getpass

def _set_env(var: str):
    """
    è®¾ç½®ç¯å¢ƒå˜é‡çš„è¾…åŠ©å‡½æ•°

    å‚æ•°:
        var (str): è¦è®¾ç½®çš„ç¯å¢ƒå˜é‡åç§°

    åŠŸèƒ½:
        - æ£€æŸ¥ç¯å¢ƒå˜é‡æ˜¯å¦å·²å­˜åœ¨
        - å¦‚æœä¸å­˜åœ¨ï¼Œåˆ™æç¤ºç”¨æˆ·è¾“å…¥å¹¶è®¾ç½®
    """
    if not os.environ.get(var):  # æ£€æŸ¥ç¯å¢ƒå˜é‡æ˜¯å¦å·²è®¾ç½®
        os.environ[var] = getpass.getpass(f"{var}: ")  # å®‰å…¨åœ°è·å–ç”¨æˆ·è¾“å…¥

# è®¾ç½® OpenAI APIä»£ç†åœ°å€ (ä¾‹å¦‚ï¼šhttps://api.apiyi.com/v1ï¼‰
# è®¾ç½® DeepSeekåœ°å€(ä¾‹å¦‚ï¼šhttps://api.deepseek.comï¼‰
_set_env("OPENAI_BASE_URL")
# è®¾ç½® OpenAI API å¯†é’¥
# è¿™æ˜¯ä½¿ç”¨ OpenAI / DeepSeek æ¨¡å‹æ‰€å¿…éœ€çš„
_set_env("OPENAI_API_KEY")
# è®¾ç½® å¤§è¯­è¨€æ¨¡å‹åç§°ï¼ˆgpt-4o / deepseek-chatï¼‰
_set_env("MODEL_NAME")
````

````python
# ===========================================
# å¯¼å…¥å¿…è¦çš„åº“å’Œæ¨¡å—
# ===========================================
import json  # ç”¨äºå¤„ç†JSONæ•°æ®æ ¼å¼
import os    # ç”¨äºæ“ä½œç³»ç»Ÿç›¸å…³æ“ä½œï¼ˆå¦‚æ–‡ä»¶åˆ—è¡¨ã€æ–‡ä»¶è¯»å–ï¼‰
from typing import List  # ç”¨äºç±»å‹æç¤ºï¼Œæé«˜ä»£ç å¯è¯»æ€§
from openai import OpenAI # ç”¨äºè°ƒç”¨OpenAI API

# ===========================================
# å·¥å…·å‡½æ•°å®šä¹‰ - æ™ºèƒ½ä½“å¯ä»¥è°ƒç”¨çš„å…·ä½“åŠŸèƒ½
# ===========================================

def list_files() -> List[str]:
    """
    åˆ—å‡ºå½“å‰ç›®å½•ä¸­çš„æ‰€æœ‰æ–‡ä»¶å’Œæ–‡ä»¶å¤¹

    è¿”å›å€¼:
        List[str]: åŒ…å«å½“å‰ç›®å½•ä¸­æ‰€æœ‰æ–‡ä»¶å’Œæ–‡ä»¶å¤¹åç§°çš„åˆ—è¡¨
    """
    return os.listdir(".")

def read_file(file_name: str) -> str:
    """
    è¯»å–æŒ‡å®šæ–‡ä»¶çš„å†…å®¹

    å‚æ•°:
        file_name (str): è¦è¯»å–çš„æ–‡ä»¶å

    è¿”å›å€¼:
        str: æ–‡ä»¶å†…å®¹ï¼ˆå¦‚æœæˆåŠŸï¼‰æˆ–é”™è¯¯ä¿¡æ¯ï¼ˆå¦‚æœå¤±è´¥ï¼‰
    """
    try:
        # å°è¯•ä»¥åªè¯»æ¨¡å¼æ‰“å¼€æ–‡ä»¶
        with open(file_name, "r") as file:
            return file.read()
    except FileNotFoundError:
        # æ–‡ä»¶ä¸å­˜åœ¨æ—¶çš„é”™è¯¯å¤„ç†
        return f"Error: {file_name} not found."
    except Exception as e:
        # å…¶ä»–å¼‚å¸¸çš„å¤„ç†ï¼ˆå¦‚æƒé™é—®é¢˜ã€æ–‡ä»¶æŸåç­‰ï¼‰
        return f"Error: {str(e)}"

def terminate(message: str) -> None:
    """
    ç»ˆæ­¢æ™ºèƒ½ä½“å¾ªç¯å¹¶æ˜¾ç¤ºæ€»ç»“æ¶ˆæ¯

    å‚æ•°:
        message (str): è¦æ˜¾ç¤ºç»™ç”¨æˆ·çš„ç»ˆæ­¢æ¶ˆæ¯
    """
    print(f"Termination message: {message}")
````

````python
# ===========================================
# å·¥å…·å‡½æ•°æ˜ å°„å­—å…¸
# ===========================================
# å°†å·¥å…·åç§°æ˜ å°„åˆ°å¯¹åº”çš„Pythonå‡½æ•°
# è¿™æ ·æ™ºèƒ½ä½“å°±å¯ä»¥é€šè¿‡å·¥å…·åç§°åŠ¨æ€è°ƒç”¨ç›¸åº”çš„å‡½æ•°
tool_functions = {
    "list_files": list_files,    # æ–‡ä»¶åˆ—è¡¨å·¥å…·
    "read_file": read_file,      # æ–‡ä»¶è¯»å–å·¥å…·
    "terminate": terminate       # ç»ˆæ­¢å·¥å…·
}

# ===========================================
# å·¥å…·é…ç½® - å®šä¹‰æ™ºèƒ½ä½“å¯ä»¥ä½¿ç”¨çš„å·¥å…·
# ===========================================
# è¿™ä¸ªé…ç½®å‘Šè¯‰å¤§è¯­è¨€æ¨¡å‹æœ‰å“ªäº›å·¥å…·å¯ä»¥ä½¿ç”¨
# ä»¥åŠæ¯ä¸ªå·¥å…·çš„å‚æ•°è¦æ±‚
tools = [
    {
        "type": "function",  # å·¥å…·ç±»å‹ï¼šå‡½æ•°è°ƒç”¨
        "function": {
            "name": "list_files",  # å·¥å…·åç§°
            "description": "Returns a list of files in the directory.",  # å·¥å…·æè¿°
            "parameters": {"type": "object", "properties": {}, "required": []}  # å‚æ•°é…ç½®ï¼ˆæ­¤å·¥å…·æ— éœ€å‚æ•°ï¼‰
        }
    },
    {
        "type": "function",
        "function": {
            "name": "read_file",
            "description": "Reads the content of a specified file in the directory.",
            "parameters": {
                "type": "object",
                "properties": {
                    "file_name": {"type": "string"}  # å®šä¹‰å‚æ•°ç±»å‹
                },
                "required": ["file_name"]  # å¿…éœ€å‚æ•°åˆ—è¡¨
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "terminate",
            "description": "Terminates the conversation. No further actions or interactions are possible after this. Prints the provided message for the user.",
            "parameters": {
                "type": "object",
                "properties": {
                    "message": {"type": "string"},  # ç»ˆæ­¢æ¶ˆæ¯å‚æ•°
                },
                "required": ["message"]  # æ¶ˆæ¯å‚æ•°æ˜¯å¿…éœ€çš„
            }
        }
    }
]
````

````python
# ===========================================
# æ™ºèƒ½ä½“è§„åˆ™å’Œç³»ç»Ÿæç¤º
# ===========================================
# å®šä¹‰æ™ºèƒ½ä½“çš„è¡Œä¸ºè§„åˆ™å’Œè§’è‰²å®šä½
agent_rules = [{
    "role": "system",  # ç³»ç»Ÿè§’è‰²ï¼Œå®šä¹‰æ™ºèƒ½ä½“çš„åŸºæœ¬è¡Œä¸º
    "content": """
You are an AI agent that can perform tasks by using available tools.

If a user asks about files, documents, or content, first list the files before reading them.

When you are done, terminate the conversation by using the "terminate" tool and I will provide the results to the user.
"""
}]
````

````python
# ===========================================
# æ™ºèƒ½ä½“å¾ªç¯å‚æ•°åˆå§‹åŒ–
# ===========================================
iterations = 0        # å½“å‰è¿­ä»£æ¬¡æ•°è®¡æ•°å™¨
max_iterations = 10   # æœ€å¤§è¿­ä»£æ¬¡æ•°é™åˆ¶ï¼Œé˜²æ­¢æ— é™å¾ªç¯

# è·å–ç”¨æˆ·ä»»åŠ¡è¾“å…¥
user_task = input("What would you like me to do? ")

# åˆå§‹åŒ–å¯¹è¯è®°å¿†ï¼Œå­˜å‚¨ç”¨æˆ·çš„ä»»åŠ¡
memory = [{"role": "user", "content": user_task}]

# ===========================================
# æ™ºèƒ½ä½“ä¸»å¾ªç¯ - æ ¸å¿ƒæ‰§è¡Œé€»è¾‘
# ===========================================
while iterations < max_iterations:

    # æ„å»ºå®Œæ•´çš„æ¶ˆæ¯åˆ—è¡¨ï¼šç³»ç»Ÿè§„åˆ™ + å¯¹è¯è®°å¿†
    messages = agent_rules + memory

    # è°ƒç”¨å¤§è¯­è¨€æ¨¡å‹ï¼Œè·å–å“åº”
    # æ¨¡å‹ä¼šæ ¹æ®ç”¨æˆ·ä»»åŠ¡å’Œå¯ç”¨å·¥å…·å†³å®šä¸‹ä¸€æ­¥è¡ŒåŠ¨
    client=OpenAI(
        base_url=os.environ['OPENAI_BASE_URL'],
        api_key=os.environ['OPENAI_API_KEY']
    )
    response = client.chat.completions.create(
         model=os.environ['MODEL_NAME'],  # æŒ‡å®šä½¿ç”¨çš„æ¨¡å‹
        messages=messages,      # å‘é€æ¶ˆæ¯å†å²
        tools=tools,           # æä¾›å¯ç”¨å·¥å…·åˆ—è¡¨
        max_tokens=1024       # é™åˆ¶å“åº”é•¿åº¦
    )

 # ===========================================
    # å¤„ç†å·¥å…·è°ƒç”¨å“åº”
    # ===========================================
    if response.choices[0].message.tool_calls:
        # å¦‚æœæ¨¡å‹å†³å®šè°ƒç”¨å·¥å…·ï¼Œæå–å·¥å…·ä¿¡æ¯
        tool = response.choices[0].message.tool_calls[0]
        tool_name = tool.function.name                    # è·å–å·¥å…·åç§°
        tool_args = json.loads(tool.function.arguments)  # è§£æå·¥å…·å‚æ•°

        # æ„å»ºåŠ¨ä½œè®°å½•ï¼Œç”¨äºè®°å¿†ç®¡ç†
        action = {
            "tool_name": tool_name,
            "args": tool_args
        }
        # ===========================================
        # å·¥å…·æ‰§è¡Œé€»è¾‘
        # ===========================================
        if tool_name == "terminate":
            # å¦‚æœæ˜¯ç»ˆæ­¢å·¥å…·ï¼Œæ˜¾ç¤ºæ¶ˆæ¯å¹¶é€€å‡ºå¾ªç¯
            print(f"Termination message: {tool_args['message']}")
            break
        elif tool_name in tool_functions:
            # å¦‚æœæ˜¯å·²çŸ¥å·¥å…·ï¼Œå°è¯•æ‰§è¡Œ
            try:
                # åŠ¨æ€è°ƒç”¨å¯¹åº”çš„å·¥å…·å‡½æ•°
                result = {"result": tool_functions[tool_name](**tool_args)}
            except Exception as e:
                # å·¥å…·æ‰§è¡Œå‡ºé”™æ—¶çš„é”™è¯¯å¤„ç†
                result = {"error":f"Error executing {tool_name}: {str(e)}"}
        else:
            # æœªçŸ¥å·¥å…·çš„é”™è¯¯å¤„ç†
            result = {"error": f"Unknown tool: {tool_name}"}

        # æ˜¾ç¤ºæ‰§è¡Œè¿‡ç¨‹å’Œç»“æœ
        print(f"Executing: {tool_name} with args {tool_args}")
        print(f"Result: {result}")

        # å°†æ‰§è¡Œç»“æœæ·»åŠ åˆ°å¯¹è¯è®°å¿†ä¸­
        # è¿™æ ·æ¨¡å‹å°±èƒ½è®°ä½ä¹‹å‰çš„æ“ä½œå’Œç»“æœ
        memory.extend([
            {"role": "assistant", "content": json.dumps(action)},  # è®°å½•æ™ºèƒ½ä½“çš„åŠ¨ä½œ
            {"role": "user", "content": json.dumps(result)}       # è®°å½•æ‰§è¡Œç»“æœ
        ])
    else:
        # å¦‚æœæ¨¡å‹æ²¡æœ‰è°ƒç”¨å·¥å…·ï¼Œç›´æ¥è¿”å›æ–‡æœ¬å“åº”
        result = response.choices[0].message.content
        print(f"Response: {result}")
        break  # ç»“æŸå¾ªç¯
````
